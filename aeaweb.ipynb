```python
# Page configuration
st.set_page_config(
    page_title="AEA Data Guidance",
    layout="wide"
)

# Sidebar
st.sidebar.title("Navigate")
page = st.sidebar.selectbox(
    "Choose a page",
    ["Home", "Guidance", "Requested Information", "Samples", "Templates", "Other"]
)

# Content
st.title("Data Guidance")

if page == "Home":
    st.write("Welcome to the Social Science Data Editors website.")
    st.warning("We are in the process of restructuring this website. Please bear with us, and report any issues.")

elif page == "Guidance":
    guidance_page = st.selectbox(
        "Select a guidance document",
        ["Data citation guide", "Additional data citation guidance", "Verification guidance", 
         "Docker guidelines", "Reproducible GIS analysis", "Journal policies", 
         "Synthetic/fake data guidance", "Licensing guidance"]
    )
    
    if guidance_page == "Data citation guide":
        st.write("Content for Data citation guide")
    elif guidance_page == "Additional data citation guidance":
        st.write("Content for Additional data citation guidance")
    # Add other guidance pages similarly

elif page == "Requested Information":
    info_page = st.selectbox(
        "Select requested information",
        ["General", "Code", "Data", "DCAS", "Hosting"]
    )
    
    if info_page == "General":
        st.write("General requested information")
    elif info_page == "Code":
        st.write("Requested information about code")
    # Add other information pages similarly

elif page == "Samples":
    sample_page = st.selectbox(
        "Select a sample",
        ["Depositing data for greater good", "Sample information data", "Sample report"]
    )
    
    if sample_page == "Depositing data for greater good":
        st.write("Sample for depositing data for greater good")
    # Add other sample pages similarly

elif page == "Templates":
    template_page = st.selectbox(
        "Select a template",
        ["README", "Inquiry about DOI by editor"]
    )
    
    if template_page == "README":
        st.write("README template")
    elif template_page == "Inquiry about DOI by editor":
        st.write("Template for inquiry about DOI by editor")

elif page == "Other":
    other_page = st.selectbox(
        "Select other information",
        ["DCAS Commercial data", "DCAS Restricted data", "Data editors guild charter"]
    )
    
    if other_page == "DCAS Commercial data":
        st.write("Information about DCAS Commercial data")
    # Add other pages similarly
```
